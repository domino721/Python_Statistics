# ch.3　Python을 이용한 데이터 분석

## 3.11 검정결과해석

### 3.11.1 p값이 0.05 이하일 때의 결과 작성법
p값이 0.05 이하일 때는 유의미한 차이가 있는 것이 됩니다.  

### 3.11.2 p값이 0.05보다 클 때의 결과 작성법
p값이 0.05보다 클 때는 귀무가설을 기각할 수 없습니다.  
p값이 0.05보다 클 때는 봉지 과자의 평균 중령은 50g과 유의미한 차이가 있다고 말할 수 없다.

### 3.11.3 가설검정을 할 때 자주 하는 실수
- p값이 작으면 차이가 크다고 말할 수 있다.
    - p값이 아무리 작다고 해도 유의미한 차이가 있는 것일 뿐이지, 차이가 크다고 말할 수 없습니다.
- p값이 0.05보다 크므로 차이가 없다고 말할 수 있다.
    - p값이 0.05보다 크다고 해서 차이가 없다고 말할수는 없다. 즉 귀무가설을 기각할 수 없다고 해서 귀무가설이 옳다곤 할 수 없는 것이다.
- 1-p 값은 대립가설이 올바를 확률이다.
    - 앞의 맥락과 같다.

### 3.11.4 제1종 오류와 제2종 오류

제 1종 오류: 귀무가설이 올바르지만 실수하여 귀무가설을 기각하는 경우  
- p값이 충분히 작으면 제 1종 오류를 범할 확률이 낮아져서 귀무가설을 기각해도 괜찮다고 판단할 수 있다.  


제 2종 오류: 귀무가설이 틀렸는데 실수하여 귀무가설을 채택하는 경우

### 3.11.5 검정의 비대칭성
검정의 비대칭성: 제2종 오류에 대해서는 컨트롤할 수 없어서 발생하는 것
- 제 2종 오류가 일어날 확률, 즉 귀무가설이 틀렸는데 실수하여 귀무가설을 채택할 확률은 통계적 가설검정으로 컨트롤할 수 없다.

통계적 가설검정에서는 제 1종 오류만 컨트롤할 수 있다는 점을 기억하자.

### 3.11.6 유의수준은 검정을 하기 전에 정해둔다.

말 그대로 검정 결과에 따라서 유의수준을 정하는 것이 아니라, 유의수준을 정하는 것이 선행된 후 검정을 해야한다.

### 3.11.7 모델선택

### 3.11.8 통계적가설검정이 필요한가

통계적 가설을 배우는 것이 훗날엔 필요하지 않을 수도 있다.
하지만 현재는 통계적 가설검정 없이 분석을 하는 것은 효율이 떨어지고 절차에 맞지 않는다.

### 3.11.9 가정이 올바른가?
t검정에서는 모집단분포가 정규분포라고 가정하였다.  
이 가정이 충족되지 않을 경우 p 값을 계산하는 것은 불가능하다.  
데이터 분석은 데이터를 다듬는 작업이라 생각할 수 있지만, 가정과의 차이를 줄이는 것이 더 중요한 과정이다.
